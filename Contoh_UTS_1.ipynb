```python
# prompt: import pandas as pd

import pandas as pd
```

```python
dataset=pd.DataFrame({
    'nim':['001','002','003'],
    'nama':['Ikhsan','Syifa','Anggita'],
    'nilai':[90,95,100],
    'prodi':['TI','BD','TI']
})
dataset
```

```python
print(f'Jumlah Data : {dataset.shape[0]}')
```

```python
print(f'Rata-rata Nilai Mahasiswa : {dataset["nilai"].mean()}')
```

```python
# Nilai Maksimum dan Minimum

print(f'Nilai Maksimum : {dataset["nilai"].max()}')
print(f'Nilai Minimum : {dataset["nilai"].min()}')
```

```python
dataset.groupby('prodi')['nim'].count()
```
