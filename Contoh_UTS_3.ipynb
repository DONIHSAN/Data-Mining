```python
# Import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

# Membuat dataset
dataset = pd.DataFrame({
    'nim': ['3122001','3122002','3122003','3122004','3122005','3122006','3122007','3122008','3122009','31220010'],
    'nama': ['Ikhsan','Syifa','Anggita','Fikri','Rani','Inay','Tiara','Hapi','Rafli','Dika'],
    'nilai': [90,95,100,70,65,55,90,88,85,93],
    'prodi': ['Teknik Informatika','Teknik Industri','Teknik Lingkungan',
              'Teknik Informatika','Teknik Industri','Teknik Lingkungan',
              'Teknik Informatika','Teknik Industri','Teknik Lingkungan','Teknik Informatika']
})

# Menampilkan dataset sebagai tabel interaktif
print("Dataset Mahasiswa:")
display(dataset)

# Statistik Deskriptif
print("\n--- Statistik Deskriptif ---")
print(f"1. Jumlah Mahasiswa: {len(dataset)}")
print(f"2. Rata-rata Nilai Mahasiswa: {dataset['nilai'].mean():.2f}")
print(f"3. Nilai Maksimum: {dataset['nilai'].max()}")
print(f"4. Nilai Minimum: {dataset['nilai'].min()}")

# Grouping: Jumlah mahasiswa tiap prodi
print("\n--- Jumlah Mahasiswa per Prodi ---")
jumlah_mahasiswa_prodi = dataset['prodi'].value_counts()
display(jumlah_mahasiswa_prodi)

# Visualisasi jumlah mahasiswa per prodi
plt.figure(figsize=(8,5))
sns.countplot(data=dataset, y='prodi', order=jumlah_mahasiswa_prodi.index, palette='viridis')
plt.title('Jumlah Mahasiswa per Prodi')
plt.xlabel('Jumlah')
plt.ylabel('Prodi')
plt.show()

# View Data

# 5. Mahasiswa Prodi Teknik Informatika
print("\n--- Data Mahasiswa Teknik Informatika ---")
ti_mahasiswa = dataset[dataset['prodi'] == 'Teknik Informatika']
display(ti_mahasiswa)

# 6. Mahasiswa dengan nilai > 85
print("\n--- Data Mahasiswa dengan Nilai > 85 ---")
nilai_diatas_85 = dataset[dataset['nilai'] > 85]
display(nilai_diatas_85)

# 7. Rata-rata nilai per prodi
print("\n--- Rata-rata Nilai per Prodi ---")
rata_rata_nilai_prodi = dataset.groupby('prodi')['nilai'].mean()
display(rata_rata_nilai_prodi)

# Visualisasi rata-rata nilai per prodi
plt.figure(figsize=(8,5))
sns.barplot(x=rata_rata_nilai_prodi.values, y=rata_rata_nilai_prodi.index, palette='coolwarm')
plt.title('Rata-rata Nilai per Prodi')
plt.xlabel('Rata-rata Nilai')
plt.ylabel('Prodi')
plt.xlim(0, 100)
plt.show()
```
